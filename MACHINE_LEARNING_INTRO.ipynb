```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train_cleaned,y_train_cleaned)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the data
car_crash = pd.read_excel('/content/new dataset.xlsx')

# Select relevant columns and drop rows with NaN values in 'Hour' or 'Latitude'
cleaned_data = car_crash[['Hour', 'Latitude']].dropna()

# Extract the independent variable (X) and dependent variable (y) from the cleaned data
X_cleaned = cleaned_data[['Hour']]
y_cleaned = cleaned_data['Latitude']

# Create and fit a linear regression model
regressor = LinearRegression()
regressor.fit(X_cleaned, y_cleaned)

# Plot the data points from the cleaned data
plt.scatter(X_cleaned, y_cleaned, marker='o', label='Data Points', alpha=0.5)

# Overlay the regression line
plt.plot(X_cleaned, regressor.predict(X_cleaned), color='red', linewidth=2, label='Regression Line')
plt.xlabel('Hour')
plt.ylabel('Latitude')
plt.title('Linear Regression of Latitude vs Hour')
plt.legend()
plt.show()
```

```python
import joblib
import pickle  # Import the pickle module
from sklearn.linear_model import LinearRegression
import pandas as pd # Import pandas to use dropna

# Combine x and y to easily drop rows with NaNs in either
train_data_to_save = pd.concat([x, y], axis=1)

# Drop rows with any NaN values
train_data_cleaned_to_save = train_data_to_save.dropna()

# Separate the cleaned data back into x and y for fitting
x_cleaned_to_save = train_data_cleaned_to_save[['Hour']]
y_cleaned_to_save = train_data_cleaned_to_save['Latitude'] # Assuming y is a Series or should be treated as such

# Train your linear regression model
model = LinearRegression()
model.fit(x_cleaned_to_save, y_cleaned_to_save)

# Save the trained model to a file using joblib
joblib.dump(model, 'linear_regression_model.pkl')

# Alternatively, you can save the model using pickle
with open('linear_regression_model.pkl', 'wb') as file:
    pickle.dump(model, file)

print("Model trained and saved successfully!")
```

```python
import joblib
import numpy as np
import pandas as pd
# Sample X values for prediction
new_X = np.array([9.8, 7, 8, 9, 10]).reshape(-1, 1)  # Reshape to a 2D array
# Load the trained model
model = joblib.load('linear_regression_model.pkl')  # Load your trained model here
# Make predictions on the new X values
predictions = model.predict(new_X)
# Create a DataFrame with one-dimensional arrays
new_data = pd.DataFrame({'X': new_X.flatten(), 'Predicted_Y': predictions.flatten()})
# Display the new_data DataFrame with the X values and the predicted Y values
print(new_data)
```
